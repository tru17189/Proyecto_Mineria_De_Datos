# ¿De qué países se están importando más vehículos? 2016

```python
#Primero Vamos a saber de cuantos paises estamos importados de este mes para comenzar
ene2016 = pd.read_csv('2016\Enero2016.csv')
ene = ene2016.groupby('pais').count()
ene2016.groupby('pais').count()
#con esta tabla podemos observar que hay una gran cantidad de paises de los que estamos importando carros, pero
#en algunos se importan una cantidad demasiado pequeña, en algunos se importa incluso un carro, asi que para este
#trabajo solo tomaremos los 5 paises de los que mas estamos importando carros en general de todos lo meses. 
```

```python
#A continuacion presentare los 5 paises de los que mas importamos de cada mes del 2016, despues guardaremos
#la cantidad total de los 5 que mas aparecieron para hacer una grafica. 
ene = ene.sort_values(by ='Marca', ascending=False)
ene.head()
```
